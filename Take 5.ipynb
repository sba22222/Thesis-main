```python
#!pip install music21

import os
from music21 import converter

# Replace 'your_root_folder' with the path to the root folder containing MIDI files
root_folder = 'POP909'

# Iterate over folders and MIDI files
for folder_name in os.listdir(root_folder):
    folder_path = os.path.join(root_folder, folder_name)

    # Check if the item in the directory is a folder
    if os.path.isdir(folder_path):
        print(f"Processing folder: {folder_name}")

        # Iterate over MIDI files in the folder
        for file_name in os.listdir(folder_path):
            if file_name.endswith('.mid'):
                midi_file_path = os.path.join(folder_path, file_name)

                # Load MIDI file
                midi_stream = converter.parse(midi_file_path)

                # Print basic information about the MIDI file
                print(f"\nMIDI file information for {file_name}:")
                print("Number of parts:", len(midi_stream.parts))

                # Iterate through the parts of the score
                for part_num, part in enumerate(midi_stream.parts):
                    print(f"\nPart {part_num + 1}:")

                    # Print time signature for the part
                    print("Time signature:", part.getTimeSignatures()[0])  # Assuming a single time signature

                    # Print key signature for the part
                    try:
                        print("Key signature:", part.flat.getElementsByClass('KeySignature')[0])
                    except IndexError:
                        print("No key signature found.")

                    # Iterate through the notes in the part
                    for element in part.flat:
                        if 'Note' in element.classes:
                            print(element)
```

```python
from music21 import converter

# Replace 'your_root_folder' with the path to the root folder containing MIDI files
root_folder = 'your_root_folder'

# Initialize an empty list to store transformed streams
transformed_streams = []

# Iterate over folders and MIDI files
for folder_name in os.listdir(root_folder):
    folder_path = os.path.join(root_folder, folder_name)

    # Check if the item in the directory is a folder
    if os.path.isdir(folder_path):
        print(f"Processing folder: {folder_name}")

        # Iterate over MIDI files in the folder
        for file_name in os.listdir(folder_path):
            if file_name.endswith('.mid'):
                midi_file_path = os.path.join(folder_path, file_name)

                # Load MIDI file
                midi_stream = converter.parse(midi_file_path)

                # Apply transformations to the midi_stream
                midi_stream.transpose('C')  # Transpose to C major
                midi_stream.getTimeSignatures()[0].numerator = 4  # Change time signature to 4/4

                # Append the transformed stream to the list
                transformed_streams.append(midi_stream)

# Now, 'transformed_streams' contains the transformed versions of each MIDI file
```
