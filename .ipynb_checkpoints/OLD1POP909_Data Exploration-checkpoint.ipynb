# Data Exploration

The below delves into the exploration and analysis of MIDI data sourced from the POP909 dataset, a collection of music tracks. The script is structured into distinct sections, each fulfilling a crucial role in understanding and processing the musical information contained within the MIDI files. Initially, the code demonstrates how to play MIDI files using the mido library, enabling auditory examination of the music. Following this, it employs the pretty_midi library to analyze MIDI data, extracting essential details such as tempo, time signature changes, and characteristics of pitched instruments. Visualizations, including histograms and scatter plots, are generated to visualize note distributions and assess pitch consistency over time. Furthermore, the code performs key signature analysis across multiple MIDI files, revealing common patterns in time signatures and expected music keys within the dataset. **Overall, this code provides a comprehensive framework for exploring MIDI music data, offering valuable insights.**

## Play One of the Songs

```python
# Define the path to the MIDI file
midi_file_path = 'POP909/909/909.mid'

# Create a MidiFile object by parsing the MIDI file located at the specified path
midi_file = MidiFile(midi_file_path)

# Open an output port for MIDI communication with the default backend
with mido.open_output() as port:
    # Iterate over each MIDI message generated by playing the MIDI file
    for message in midi_file.play():
        # Send each MIDI message to the output port
        port.send(message)
```

```python
# Define a function to extract notes from a MIDI part
def extract_notes(midi_part):
    parent_element = []  # List to store parent elements of notes or chords
    ret = []  # List to store pitch values (in MIDI note numbers) of notes or chords
    for nt in midi_part:  # Iterate over notes or chords in the MIDI part
        if isinstance(nt, note.Note):  # Check if the element is a single note
            ret.append(max(0.0, nt.pitch.ps))  # Add the MIDI note number to the list, ensuring it's non-negative
            parent_element.append(nt)  # Append the note to the list of parent elements
        elif isinstance(nt, chord.Chord):  # Check if the element is a chord
            for pitch in nt.pitches:  # Iterate over pitches in the chord
                ret.append(max(0.0, pitch.ps))  # Add the MIDI note number of each pitch to the list, ensuring it's non-negative
                parent_element.append(nt)  # Append the chord to the list of parent elements
    return ret, parent_element  # Return the list of pitch values and the list of parent elements

# Define a function to print the contour of MIDI notes in a stream
def print_parts_countour(midi_stream, start_measure, end_measure):
    fig = plt.figure(figsize=(12, 5))  # Create a figure for the plot with a specified size
    ax = fig.add_subplot(1, 1, 1)  # Add a subplot to the figure
    minPitch = pitch.Pitch('C10').ps  # Initialize the minimum pitch value as MIDI note number of C10
    maxPitch = 0  # Initialize the maximum pitch value as 0
    xMax = 0  # Initialize the maximum value of x-axis as 0
    
    # Drawing notes.
    for i in range(len(midi_stream.parts)):  # Iterate over parts in the MIDI stream
        top = [n for n in midi_stream.parts[i].flat.notes if start_measure <= n.offset < end_measure]  # Filter notes in the specified measures
        y, parent_element = extract_notes(top)  # Extract pitch values and parent elements from the filtered notes
        if len(y) < 1:  # If there are no pitch values, continue to the next iteration
            continue
            
        x = [n.offset for n in parent_element]  # Get the offsets of parent elements as x values
        ax.scatter(x, y, alpha=0.6, s=7)  # Scatter plot of notes with specified alpha and size
        
        aux = min(y)  # Get the minimum pitch value
        if aux < minPitch:  # Update the minimum pitch value if necessary
            minPitch = aux
            
        aux = max(y)  # Get the maximum pitch value
        if aux > maxPitch:  # Update the maximum pitch value if necessary
            maxPitch = aux
            
        aux = max(x)  # Get the maximum x value
        if aux > xMax:  # Update the maximum x value if necessary
            xMax = aux
    
    # Draw red lines representing each octave
    for i in range(1, 10):  # Iterate over octaves from 1 to 9
        linePitch = pitch.Pitch('C{0}'.format(i)).ps  # Calculate the MIDI note number of C in the current octave
        if minPitch < linePitch < maxPitch:  # Check if the octave is within the pitch range
            ax.add_line(mlines.Line2D([0, xMax], [linePitch, linePitch], color='red', alpha=0.1))  # Add a red line representing the octave

# Parse the MIDI file and store it in a variable named midi_data
midi_data = converter.parse('C:/Users/naomi/Thesis/Thesis/Thesis-main/POP909/909/909.mid')
```

We can take a look on the pitch histogram to see which notes are most used.

```python
# Plot a histogram of pitch classes from the MIDI data
midi_data.plot('histogram', 'pitchClass', 'count');
```

The histogram allows us to visualize the distribution of pitches across different pitch classes in the MIDI file. Each bin on the x-axis represents a specific pitch class, such as C, C sharp, D etc. The height of each bar (count on the y-axis) indicates the frequency of occurrence of notes belonging to that pitch class. This distribution helps us understand the predominant pitches or musical keys present in the composition.

By examining the distribution of pitch classes, we can identify any melodic tendencies or patterns within the music. For instance, a higher frequency of certain pitch classes may indicate recurring motifs or melodies (e.g. c and G), while less frequent pitch classes may represent occasional embellishments or variations (e.g. C sharp, Eb, F sharp).

Analyzing the histogram can aid in identifying the key signature of the piece. In tonal music, certain pitch classes are more prevalent than others based on the key signature. For example, in the key of C major, pitches like C, D, E, F, G, A, and B are more likely to occur frequently, leading to higher counts in their respective bins on the histogram.

The scatter plot below shows that the use of notes look consistent through time, so there are no key changes in this piece.

```python
# Plot a scatter plot of MIDI note pitches against their offsets in the MIDI data
midi_data.plot('scatter', 'offset', 'pitchClass');
```

The scatter plot allows us to visualize the melodic patterns present in the MIDI file. Each point on the plot represents a note, with its position indicating both the pitch class (y-axis) and the measure number (x-axis) where the note occurs. Clusters of points or recurring patterns can indicate melodic motifs, phrases, or repetitions within the composition.

By observing the range of pitches represented on the y-axis, we can discern the overall pitch range of the piece. A wider spread of points suggests a broader pitch range, while a narrower spread indicates a more limited range. This information is valuable for understanding the melodic contour and expressive range of the music.

Examining the distribution and density of points across different measure numbers provides insights into the structural organization of the composition. Patterns such as repetition, variation, climax points, or transitions between sections can be identified based on the arrangement of notes on the scatter plot.

```python
# Extract the time signature information from the MIDI data
timeSignature = midi_data.getTimeSignatures()[0]

# Analyze the key of the music using the MIDI data
music_analysis = midi_data.analyze('key')

# Print the time signature of the music
print("Music time signature: {0}/{1}".format(timeSignature.beatCount, timeSignature.denominator))

# Print the expected key of the music
print("Expected music key: {0}".format(music_analysis))

# Print the confidence level of the expected key
print("Music key confidence: {0}".format(music_analysis.correlationCoefficient))

# Print other potential key alternatives with confidence levels higher than 0.5
print("Other music key alternatives:")
for analysis in music_analysis.alternateInterpretations:
    if (analysis.correlationCoefficient > 0.5):
        print(analysis)
```

```python
def analyze_midi_file(file_path):
    try:
        # Load MIDI file
        midi_data = converter.parse(file_path)

        # Extract key signature
        key_signature = midi_data.analyze('key')

        # Return key signature
        return key_signature

    except Exception as e:
        return None

def analyze_midi_folder(root_folder):
    keys_found = []  # List to store all found key signatures

    for root, dirs, files in os.walk(root_folder):
        for file in files:
            if file.lower().endswith(".mid") or file.lower().endswith(".midi"):
                file_path = os.path.join(root, file)
                key_signature = analyze_midi_file(file_path)
                if key_signature:
                    keys_found.append(str(key_signature))  # Append key signature to the list

    return keys_found

# Analyze MIDI files in the POP909 folder and capture key signatures
key_signatures = analyze_midi_folder('POP909')
```

```python
# Plot histogram of key signatures
plt.figure(figsize=(10, 6))
plt.hist(key_signatures, edgecolor='black')
plt.xlabel('Key Signature')
plt.ylabel('Frequency')
plt.title('Distribution of Expected Music Keys')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
