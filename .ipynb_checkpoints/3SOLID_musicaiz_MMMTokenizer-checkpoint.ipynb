```python
import os
from pathlib import Path
from musicaiz.tokenizers import MMMTokenizer, MMMTokenizerArguments

# Input and output directories
input_directory = r"C:\Users\naomi\Thesis\Thesis\Thesis-main\output\POP909-chunked"
output_directory = r"C:\Users\naomi\Thesis\Thesis\Thesis-main\tokenized_output"

# Tokenization parameters
args = MMMTokenizerArguments(
    windowing=True,
    time_unit="SIXTEENTH",
    num_programs=None,
    shuffle_tracks=True,
    track_density=False,
    window_size=4,
    hop_length=1,
    time_sig=False,
    velocity=False,
    quantize=False,
    tempo=True,
)

# Iterate through MIDI files in the input directory
for root, dirs, files in os.walk(input_directory):
    for file in files:
        if file.endswith(".mid"):
            midi_path = os.path.join(root, file)

            try:
                # Tokenize file
                tokenizer = MMMTokenizer(midi_path, args)
                tokenized_output = tokenizer.tokenize_file()

                # Save tokenized output
                output_path = os.path.join(output_directory, f"{file.replace('.mid', '_token-sequences.txt')}")
                with open(output_path, 'w') as output_file:
                    output_file.write(tokenized_output)

            except Exception as e:
                print(f"Error processing {midi_path}: {e}")
                continue

# Get vocabulary
files = [file for file in Path(output_directory).rglob("*_token-sequences.txt")]

if len(files) == 0:
    raise ValueError(f"Tokens sequences files not found in {output_directory}. Could not process tokens.")

vocabulary = []
for file in files:
    with open(file, 'r') as f:
        content = f.read()
        vocabulary.extend(content.split())

# Remove duplicates
vocabulary = list(set(vocabulary))

# Save vocabulary
vocabulary_path = os.path.join(output_directory, "vocabulary.txt")
with open(vocabulary_path, 'w') as vocab_file:
    for word in vocabulary:
        vocab_file.write(f"{word}\n")
```

### I want to put it all into one dataset so I have concated each song into its only row of tokenized words below

```python
import os
from pathlib import Path
from musicaiz.tokenizers import MMMTokenizer, MMMTokenizerArguments

# Input and output directories
input_directory = r"C:\Users\naomi\Thesis\Thesis\Thesis-main\output\POP909-chunked"
output_directory = r"C:\Users\naomi\Thesis\Thesis\Thesis-main\tokenized_output_v2"

# Tokenization parameters
args = MMMTokenizerArguments(
    windowing=True,
    time_unit="SIXTEENTH",
    num_programs=None,
    shuffle_tracks=True,
    track_density=False,
    window_size=4,
    hop_length=1,
    time_sig=False,
    velocity=False,
    quantize=False,
    tempo=True,
)

# Initialize variables to store tokenized outputs and vocabulary
all_tokenized_outputs = []
vocabulary = set()

# Iterate through MIDI files in the input directory
for root, dirs, files in os.walk(input_directory):
    for file in files:
        if file.endswith(".mid"):
            midi_path = os.path.join(root, file)

            try:
                # Tokenize file
                tokenizer = MMMTokenizer(midi_path, args)
                tokenized_output = tokenizer.tokenize_file()

                # Append tokenized output to the list
                all_tokenized_outputs.append(tokenized_output)

                # Update vocabulary set
                vocabulary.update(tokenized_output.split())

            except Exception as e:
                print(f"Error processing {midi_path}: {e}")
                continue

# Save all tokenized outputs into one file
all_tokenized_path = os.path.join(output_directory, "all_tokenized_outputs.txt")
with open(all_tokenized_path, 'w') as all_tokenized_file:
    all_tokenized_file.write('\n'.join(all_tokenized_outputs))

# Save vocabulary
vocabulary_path = os.path.join(output_directory, "vocabulary.txt")
with open(vocabulary_path, 'w') as vocab_file:
    for word in vocabulary:
        vocab_file.write(f"{word}\n")
```
